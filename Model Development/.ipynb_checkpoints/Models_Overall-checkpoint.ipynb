```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, plot_confusion_matrix, f1_score, accuracy_score, precision_score, recall_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.svm import SVC

try:
    import cPickle as pickle
except:
    import pickle
    
import joblib

import warnings
warnings.filterwarnings('ignore')
```

```python
def model_save(model, pickle_file):
    """Save compiled models for reuse."""
    
    with open(pickle_file, 'wb') as f:
        pickle.dump(model, f)

def model_load(file):
    """Reload compiled models for reuse."""
    
    with open(file, 'rb') as f:
        clf = joblib.load(f)
        
    return clf
```

```python
# read in data

raw = pd.read_csv('../Data/dat_raw.csv', index_col=0)
X_train = pd.read_csv('../Data/X_train.csv', index_col=0)
X_test = pd.read_csv('../Data/X_test.csv', index_col=0)
y_train = pd.read_csv('../Data/y_train.csv', index_col=0).squeeze()
y_test = pd.read_csv('../Data/y_test.csv', index_col=0).squeeze()
```

```python
# scaled data, minmax
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# # scaled data, normal
# scaler2 = StandardScaler()
# X_train_scaled2 = scaler2.fit_transform(X_train)
# X_test_scaled2 = scaler2.transform(X_test)
```

## Model Performance

```python
# dummy, same label distribution in train & test

# dummy_clf = DummyClassifier(strategy="most_frequent")
# dummy_clf.fit(X_train, y_train)
dummy_clf = model_load('dummy.pickle')

train_pred_dummy = dummy_clf.predict(X_train)
test_pred_dummy = dummy_clf.predict(X_test)

print('Training')
print(classification_report(y_train, train_pred_dummy, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

print('Testing')
print(classification_report(y_test, test_pred_dummy, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

disp_dummy = plot_confusion_matrix(dummy_clf, X_test, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize='true')
```

```python
# kNN, did normalization: 0.67 train, 0.62 test, snf 0.17
# did standardization: 0.65 train, 0.60 test, snf 0.2, picked this one

# knn_clf = KNeighborsClassifier(n_neighbors=10, leaf_size=50, weights='uniform', p=1)
# knn_clf.fit(X_train_scaled, y_train)
knn_clf = model_load('knn.pickle')

train_pred_knn = knn_clf.predict(X_train_scaled)
test_pred_knn = knn_clf.predict(X_test_scaled)

print('Training')
print(classification_report(y_train, train_pred_knn, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

print('Testing')
print(classification_report(y_test, test_pred_knn, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

disp_knn = plot_confusion_matrix(knn_clf, X_test_scaled, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize='true')
```

```python
# logreg, L1 regularization with feature selection (L2 doesn't perform well), did standardization, ovr

# logreg_clf = LogisticRegression(penalty='l1', solver='liblinear', multi_class='ovr', random_state=0)
# logreg_clf.fit(X_train_scaled, y_train)
logreg_clf = model_load('logreg.pickle')

train_pred_logreg = logreg_clf.predict(X_train_scaled)
test_pred_logreg = logreg_clf.predict(X_test_scaled)

print('Training')
print(classification_report(y_train, train_pred_logreg, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

print('Testing')
print(classification_report(y_test, test_pred_logreg, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

disp_logreg = plot_confusion_matrix(logreg_clf, X_test_scaled, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize='true')
```

```python
# SVM, did standardization, ovo

# svm_clf = SVC(C=1.0, decision_function_shape='ovo', degree=5, kernel='poly')
# svm_clf.fit(X_train_scaled, y_train)
svm_clf = model_load('svm.pickle')

train_pred_svm = svm_clf.predict(X_train_scaled)
test_pred_svm = svm_clf.predict(X_test_scaled)

print('Training')
print(classification_report(y_train, train_pred_svm, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

print('Testing')
print(classification_report(y_test, test_pred_svm, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

disp_svm = plot_confusion_matrix(svm_clf, X_test_scaled, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize='true')
```

```python
# XGBoost

# param:
# XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
#               colsample_bynode=1, colsample_bytree=0.266798290436423,
#               gamma=3.2261552797199973, gpu_id=-1, importance_type='gain',
#               interaction_constraints='', learning_rate=0.04372790493554727,
#               max_delta_step=0, max_depth=6, min_child_weight=1, missing=nan,
#               monotone_constraints='()', n_estimators=887, n_jobs=0,
#               num_parallel_tree=1, objective='multi:softprob', random_state=0,
#               reg_alpha=0, reg_lambda=1, scale_pos_weight=None, subsample=1,
#               tree_method='exact', validate_parameters=1, verbosity=None)

# readin model
xgb_clf = model_load('gbm_gamma.pickle')
    
train_pred_xgb = xgb_clf.predict(X_train)
test_pred_xgb = xgb_clf.predict(X_test)

print('Training')
print(classification_report(y_train, train_pred_xgb, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

print('Testing')
print(classification_report(y_test, test_pred_xgb, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

disp_xgb = plot_confusion_matrix(xgb_clf, X_test, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize='true')
```

```python
# RF

# param:
# RandomForestClassifier(bootstrap=False, max_depth=50, max_features='sqrt',
#                        min_samples_leaf=2, min_samples_split=10,
#                        n_estimators=1100)

# readin model
import zipfile
with zipfile.ZipFile('rf_pickle.zip', 'r') as zip_ref:
    zip_ref.extractall()

rf_clf = model_load('rf.pickle')

train_pred_xgb = rf_clf.predict(X_train)
test_pred_xgb = rf_clf.predict(X_test)

print('Training')
print(classification_report(y_train, train_pred_xgb, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

print('Testing')
print(classification_report(y_test, test_pred_xgb, target_names=['HOME', 'SNF', 'Others', 'Dead/Expired']))

disp_rf = plot_confusion_matrix(rf_clf, X_test, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize='true')
```
